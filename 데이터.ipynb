```python
# prompt: /content/vgsales.csv 로부터 세일을 많이한 게임 상위 10가지와 세일과 구매량의 상관관계를 그래프로 만들어줘

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_csv('/content/vgsales.csv')

# 판매량 상위 10개 게임을 추출합니다.
top_10_games = df.nlargest(10, 'Global_Sales')

# 상위 10개 게임의 이름과 글로벌 판매량을 출력합니다.
print("판매량 상위 10개 게임:")
print(top_10_games[['Name', 'Global_Sales']])


# 'Global_Sales'와 'NA_Sales' 간의 상관관계를 계산합니다.
correlation = df['Global_Sales'].corr(df['NA_Sales'])
print(f"\nGlobal Sales와 NA_Sales의 상관관계: {correlation}")


# 'Global_Sales'와 'NA_Sales'의 관계를 산점도로 표시합니다.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Global_Sales', y='NA_Sales', data=df)
plt.title('Global Sales vs. NA Sales')
plt.xlabel('Global Sales')
plt.ylabel('NA Sales')
plt.show()

# 상위 10개 게임의 글로벌 판매량을 막대 그래프로 표시합니다.
plt.figure(figsize=(12, 6))
plt.bar(top_10_games['Name'], top_10_games['Global_Sales'])
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Best Selling Games')
plt.xlabel('Game Name')
plt.ylabel('Global Sales')
plt.tight_layout()
plt.show()
```
